```python
#!/usr/bin/env python3

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#alpha = 1.5

b = np.array([[1],[1]])

A = np.array([[1, 0.1], [0.1, 1]])

rounds = 1000

def get_sample():

	r = np.random.randint(2)

	if r == 0:
		return -1
	else:
		return 1


def main():
	
	#theta(zero)
	theta = np.array([[0], [0]])
	theta_values = []
	theta_values.append(theta)

	t_axis = []
	t_axis.append(0)

	for t in range(1, rounds):

		temp = np.copy(theta)
		temp = theta_values[t-1] + (0.1/(100+t))*(b - np.dot(A, theta_values[t-1]) + get_sample())
		#print(temp)
		theta_values.append(temp)
		t_axis.append(t)


	fig=plt.figure(figsize=(10,6))
	ax = fig.add_subplot(111, projection='3d')

	x_axis = []
	y_axis = []

	for i in range(0, rounds):
		x_axis.append(theta_values[i][0,0])
		y_axis.append(theta_values[i][1,0])

	ax.plot(x_axis, y_axis, t_axis)
	plt.show()


if __name__ == '__main__':
	main()
```

```python
b = np.array([[1],[-1]])

A = np.array([[2, 0], [1, 0]])

main()
```
