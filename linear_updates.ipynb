```python
#!/usr/bin/env python3

import numpy as np
import matplotlib.pyplot as plt
import math

def get_sample():

	r = np.random.randint(2)

	if r == 0:
		return -1
	else:
		return 1

def get_t1(x, t, n):
	
	return x[t-1] + (1/(t))*(n - x[t-1])

def get_t2(x, t, n):

	return x[t-1] + (1/(t+100))*(n - x[t-1])

def get_t(x, t, n, alpha):

	return x[t-1] + (alpha)*(n - x[t-1])


if __name__ == '__main__':
		
		avg = []
		avg_value = 0
		sum_value = 0
		x_axis = []

		l1 = []
		l2 = []

		t1 = []       # 1/t+0
		t1.append(0)

		t2 = []		  #1/t+100
		t2.append(0)

		a1 = []		  #const alpha = 1
		a1.append(0)

		a2 = []		  #const alpha = 0.1
		a2.append(0)

		a3 = []		  #const alpha = 0.01
		a3.append(0)



		for t in range(1, 1001):

			n = get_sample()
			sum_value = sum_value + n
			avg_value = (sum_value/t)

			l1.append(1/math.sqrt(t))
			l2.append(-(1/math.sqrt(t)))

			avg.append(avg_value)
			x_axis.append(t)

			if t == 1:
				t1.append(5)
				t2.append(5)
				a1.append(0)
				a2.append(0)
				a3.append(0)

			else:
				t1.append(get_t1(t1, t, n))
				t2.append(get_t2(t2, t, n))
				a1.append(get_t(a1, t, n, 0.001))
				a2.append(get_t(a2, t, n, 0.1))
				a3.append(get_t(a3, t, n, 0.01))


		plt.plot(x_axis, avg)
		plt.plot(x_axis, l1)
		plt.plot(x_axis, l2)

		plt.show()
```

```python
		plt.plot(x_axis, t1[1:1001],label = '1/t')
		plt.plot(x_axis, t2[1:1001],label ='1/t+100')
		plt.legend()
```
